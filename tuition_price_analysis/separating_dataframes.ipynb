```python
#import dependencies
from pathlib import Path
import pandas as pd
```

```python
#import tution.csv
#changed the name of the dataframe to tuition_df -LC
tuition_df=pd.read_csv(
    Path('./Resources/tuition.csv')
)
tuition_df.head(110)        
```

```python
#keep only the columns for total tuition, fees, and room and board in 2020-2021 dollars
tuition_df=(tuition_df[['year','All insti-tutions','4-year','2-year']])
display(tuition_df.head())
display(tuition_df.tail())
```

```python
#rename columns
columns = ['Year', 'All Institutions', '4-year', '2-year']
tuition_df.columns=columns
tuition_df.head()
```

```python
#remove currency symbols from the dataframe
tuition_df.loc[:, 'All Institutions'] = tuition_df.loc[:, 'All Institutions'].str.replace("$", "")
tuition_df.loc[:, '4-year'] = tuition_df.loc[:, '4-year'].str.replace("$", "")
tuition_df.loc[:, '2-year'] = tuition_df.loc[:, '2-year'].str.replace("$", "")
tuition_df.head()
```

```python
#remove commas from the dataframe
tuition_df.loc[:, 'All Institutions'] = tuition_df.loc[:, 'All Institutions'].str.replace(",", "")
tuition_df.loc[:, '4-year'] = tuition_df.loc[:, '4-year'].str.replace(",", "")
tuition_df.loc[:, '2-year'] = tuition_df.loc[:, '2-year'].str.replace(",", "")
tuition_df.head()
```

```python
#create all institutions dataframe
all_institutions_df = tuition_df.loc[0:53, :]
#write the dataframe to a CSV file
all_institutions_path = Path('./Resources/all_institutions.csv')
all_institutions_df.to_csv(all_institutions_path, index=False)
display(all_institutions_df.head())
display(all_institutions_df.tail())
```

```python
#create public institutions dataframe
public_df = tuition_df.loc[55:108, :]
#write the dataframe to a CSV file
public_df_path = Path('./Resources/public_institutions.csv')
public_df.to_csv(public_df_path, index=False)
display(public_df.head())
display(public_df.tail())
```

```python
#create all private institutions dataframe
#since the separated for-profit and nonprofit sections only contain data from 1999-2021, we will only use the data for all private institutions combined.
all_private_df = tuition_df.loc[110:163, :]
#write the dataframe to a CSV file
private_df_path = Path('./Resources/all_private_institutions.csv')
all_private_df.to_csv(private_df_path, index=False)
display(all_private_df.head())
display(all_private_df.tail())
```
